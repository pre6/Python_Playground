# Spirographs are cool

```python

import math
import random
import matplotlib.pyplot as plt

R = random.randint(0,360)
r = random.randint(0,360)
p_c = random.randint(0,360)

print(R,r,p_c)

def spirograph(R,r,p_c):

    divisor = math.gcd(R,r)
    repeat = int(r/divisor)

    x = []
    y= [] 


    for i in range(0,360*repeat):
        angle = math.radians(i)
        x.append((R-r)*math.cos(angle) + p_c*math.cos(((R-r)/r)*angle))
        y.append((R-r)*math.sin(angle) - p_c*math.sin(((R-r)/r)*angle))

    
    
    return [x, y]
fig = plt.figure(figsize=(10, 10))   
plt.plot(spirograph(R,r,p_c)[0],spirograph(R,r,p_c)[1])
```

# Categorizing the different shapes of Spirographs Based on the inputs for R, r and p_c
